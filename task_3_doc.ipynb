While these types donâ€™t have as many built-in methods as strings or lists, they do support some useful ones, especially for `int` and `complex`. Here's a quick guide:


### ğŸ”¢ Common Functions for All Number Types

- `sum(iterable)` â€“ Adds up all elements in an iterable  


### ğŸ§  Integer (`int`) Methods

```python
x = 10
print(x.bit_length())  # Output: 4
```


### ğŸŒŠ Float (`float`) Specifics
Floats donâ€™t have custom methods, but they work with:
- `is_integer()` â€“ Checks if the float is a whole number  
```python
x = 5.0
print(x.is_integer())  # Output: True
```


### ğŸŒ€ Complex Number Methods
- `.real` â€“ Returns the real part

## ___List Methods___

Python lists come with a powerful set of **built-in methods** that let you manipulate, search, and organize data with ease. Here's a handy rundown:

---

### ğŸ§° Common List Methods

| Method         | Description |
|----------------|-------------|
| `append(x)`    | Adds `x` to the end of the list |
| `extend(iter)` | Adds all elements from an iterable (like another list) |
| `insert(i, x)` | Inserts `x` at index `i` |
| `remove(x)`    | Removes the first occurrence of `x` |
| `pop(i)`       | Removes and returns the item at index `i` (last item if `i` is not given) |
| `clear()`      | Removes all items from the list |
| `index(x)`     | Returns the index of the first occurrence of `x` |
| `count(x)`     | Returns the number of times `x` appears |
| `sort()`       | Sorts the list in ascending order (can take `key` and `reverse`) |
| `reverse()`    | Reverses the list in place |
| `copy()`       | Returns a shallow copy of the list |


### ğŸ§ª Example in Action

```python
fruits = ['apple', 'banana', 'cherry']
fruits.append('date')
fruits.sort()
print(fruits)  # Output: ['apple', 'banana', 'cherry', 'date']
```
